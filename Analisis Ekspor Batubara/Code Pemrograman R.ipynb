# Cek Outlier, Extreme Value, dan Influential

```r
library(ggplot2)
library(dplyr)
library(tidyr)

get_bounds <- function(x, multiplier) {
  Q1 <- quantile(x, 0.25, na.rm = TRUE)
  Q3 <- quantile(x, 0.75, na.rm = TRUE)
  IQR_val <- Q3 - Q1
  lower <- Q1 - multiplier * IQR_val
  upper <- Q3 + multiplier * IQR_val
  return(c(lower = lower, upper = upper))
}

bounds_1_5 <- get_bounds(df$fobvalue, 1.5)
bounds_3 <- get_bounds(df$fobvalue, 3)


df <- df %>%
  mutate(
    outlier_1_5 = ifelse(fobvalue < bounds_1_5["lower"] | fobvalue > bounds_1_5["upper"], TRUE, FALSE),
    outlier_3 = ifelse(fobvalue < bounds_3["lower"] | fobvalue > bounds_3["upper"], TRUE, FALSE)
  )

df_long <- df %>%
  select(fobvalue, outlier_1_5, outlier_3) %>%
  pivot_longer(cols = starts_with("outlier"), names_to = "type", values_to = "is_outlier") %>%
  mutate(type = recode(type, outlier_1_5 = "1.5 x IQR", outlier_3 = "3 x IQR"))


bounds_df <- data.frame(
  type = c("1.5 x IQR", "1.5 x IQR", "3 x IQR", "3 x IQR"),
  bound = c(bounds_1_5["lower"], bounds_1_5["upper"], bounds_3["lower"], bounds_3["upper"]),
  bound_type = rep(c("Lower", "Upper"), 2),
  color = rep(c("red", "blue"), each = 2)
)


ggplot(df_long, aes(x = type, y = fobvalue)) +
  geom_boxplot(outlier.shape = NA) +
  geom_jitter(aes(color = is_outlier), width = 0.15, alpha = 0.6, size = 1.7) +
  geom_hline(data = bounds_df, aes(yintercept = bound, color = color), linetype = "dashed", size = 1) +
  scale_color_manual(values = c("FALSE" = "black", "TRUE" = "blue"), guide = "legend") +
  scale_y_continuous(name = "fobvalue") +
  labs(title = "Boxplot Outlier fobvalue dengan Batas 1.5 dan 3 x IQR",
       x = "Batas Outlier",
       color = "Outlier") +
  theme_minimal() +
  theme(legend.position = "bottom") +
  guides(color = guide_legend(override.aes = list(size=3)))
```

```r
num_vars <- names(df)[sapply(df, is.numeric)]

for (var in num_vars) {

  Q1 <- quantile(df[[var]], 0.25, na.rm = TRUE)
  Q3 <- quantile(df[[var]], 0.75, na.rm = TRUE)

  IQR_value <- Q3 - Q1

  lower_bound <- Q1 - 1.5 * IQR_value
  upper_bound <- Q3 + 1.5 * IQR_value

  outlier_values <- df[[var]][df[[var]] < lower_bound | df[[var]] > upper_bound]

  cat("OUTLIER UNTUK VARIABEL:", var, "\n")
  cat("Lower bound:", lower_bound, "\n")
  cat("Upper bound:", upper_bound, "\n")

  if (length(outlier_values) == 0) {
    cat("Tidak ada outlier.\n")
  } else {
    cat("Outlier values:\n")
    print(outlier_values)
  }
}
```

# Penanganan Outlier, Extreme Value, dan Influential

```r
df1 <- df

safe_log <- function(x) {
  if (all(x <= 0, na.rm = TRUE)) {
    stop("Semua nilai <= 0, log tidak bisa dihitung!")
  }
  min_pos <- min(x[x > 0], na.rm = TRUE)
  x[x <= 0] <- 0.5 * min_pos
  return(log(x))
}

winsor_both <- function(x, p = 0.05) {
  if (all(is.na(x))) return(x)
  q_low  <- quantile(x, probs = p, na.rm = TRUE)
  q_high <- quantile(x, probs = 1 - p, na.rm = TRUE)
  x[x < q_low]  <- q_low
  x[x > q_high] <- q_high
  return(x)
}

df1$fob_clean   <- winsor_both(df1$fobvalue, p = 0.05)
df1$ln_fob      <- safe_log(df1$fob_clean)
df1$ln_GDPj     <- safe_log(df1$Nilai_GDP)
df1$ln_GDPI     <- safe_log(df1$GDPI)
df1$ln_dist     <- safe_log(df1$dist)
df1$ln_Pop      <- safe_log(df1$Populasi)
df1$ln_tukar    <- safe_log(df1$nilai_tukar_rill)
```

```r
num_vars <- names(df1)[sapply(df1, is.numeric)]

for (var in num_vars) {

  Q1 <- quantile(df1[[var]], 0.25, na.rm = TRUE)
  Q3 <- quantile(df1[[var]], 0.75, na.rm = TRUE)

  IQR_value <- Q3 - Q1

  lower_bound <- Q1 - 3 * IQR_value
  upper_bound <- Q3 + 3 * IQR_value

  outlier_values <- df1[[var]][df1[[var]] < lower_bound | df1[[var]] > upper_bound]

  cat("EXTREME VALUE UNTUK VARIABEL:", var, "\n")
  cat("Lower bound:", lower_bound, "\n")
  cat("Upper bound:", upper_bound, "\n")

  if (length(outlier_values) == 0) {
    cat("Tidak ada outlier.\n")
  } else {
    cat("Outlier values:\n")
    print(outlier_values)
  }
}
```

```r
num_vars <- names(df1)[sapply(df1, is.numeric)]

for (var in num_vars) {

  Q1 <- quantile(df1[[var]], 0.25, na.rm = TRUE)
  Q3 <- quantile(df1[[var]], 0.75, na.rm = TRUE)

  IQR_value <- Q3 - Q1

  lower_bound <- Q1 - 1.5 * IQR_value
  upper_bound <- Q3 + 1.5 * IQR_value

  outlier_values <- df1[[var]][df1[[var]] < lower_bound | df1[[var]] > upper_bound]

  cat("EXTREME VALUE UNTUK VARIABEL:", var, "\n")
  cat("Lower bound:", lower_bound, "\n")
  cat("Upper bound:", upper_bound, "\n")

  if (length(outlier_values) == 0) {
    cat("Tidak ada outlier.\n")
  } else {
    cat("Outlier values:\n")
    print(outlier_values)
  }
}
```

```r
lm_model <- lm(
  ln_fob_s ~ ln_GDPj_s + ln_GDPI_s + ln_dist_s + ln_Pop_s + contig + ln_tukar_s
  + fta + WTO + Trade_Freedom + Investment_Freedom +
    Business_Freedom + Monetary_Freedom + Financial_Freedom,
  data = df1
)
cooks_d <- cooks.distance(lm_model)
threshold <- 4 / length(cooks_d)
influential_idx <- which(cooks_d > threshold)

cat("\nObservasi berpengaruh (Cook’s D > 4/n):\n")
print(influential_idx)
```

```r
install.packages("frontier")
library(plm)
library(frontier)
library(dplyr)

panel <- pdata.frame(df1, c("partnerDesc", "period"))

winsor_both <- function(x, p = 0.05) {
  if (all(is.na(x))) return(x)
  q_low  <- quantile(x, probs = p, na.rm = TRUE)
  q_high <- quantile(x, probs = 1 - p, na.rm = TRUE)
  x[x < q_low]  <- q_low
  x[x > q_high] <- q_high
  return(x)
}

panel$fob_w1 <- winsor_both(panel$fobvalue, p = 0.05)
panel$fob_pc <- panel$fob_w1
panel$ln_fob_pc <- log(panel$fob_pc + 1e-6)
panel$Populasi_w1 <- winsor_both(panel$Populasi, p = 0.05)
panel$nilai_tukar_rill_w1 <- winsor_both(panel$nilai_tukar_rill, p = 0.05)
sfa_model <- sfa(
  ln_fob_pc ~ log(Nilai_GDP) + log(GDPI) + log(dist) +
    log(Populasi) + log(nilai_tukar_rill) + contig |
    fta + WTO + Monetary_Freedom + Financial_Freedom +
    Business_Freedom + Trade_Freedom + Investment_Freedom,
  data = panel,
  ineffDecrease = TRUE,
  truncNorm = TRUE,
  timeEffect = TRUE
)


summary(sfa_model)
```

```r
library(plm)

pooling <- plm(
     ln_fob_pc ~ log(Nilai_GDP) + log(GDPI) + log(dist) + log(Populasi) + log(nilai_tukar_rill) + contig +
    fta + WTO + Monetary_Freedom + Financial_Freedom + Business_Freedom + Trade_Freedom + Investment_Freedom,
  data = panel,
  model = "pooling"
)
summary(pooling)

fixed <- plm(ln_fob_pc ~ log(Nilai_GDP) + log(GDPI) + log(dist) + log(Populasi) + log(nilai_tukar_rill) + contig+
    fta + WTO + Monetary_Freedom + Financial_Freedom + Business_Freedom + Trade_Freedom + Investment_Freedom,
  data = panel, model = "within")
summary(fixed)

chow_test <- pFtest(fixed, pooling)
print(chow_test)
```

```r
library(plm)
library(lmtest)
library(sandwich)

random <- plm(
  ln_fob_pc ~ log(Nilai_GDP) + log(GDPI) + log(dist) + log(Populasi) +
    log(nilai_tukar_rill) + contig + fta + WTO +
    Monetary_Freedom + Financial_Freedom + Business_Freedom +
    Trade_Freedom + Investment_Freedom,
  data  = panel,
  model = "random"
)


vcov_cluster <- vcovHC(
  random,
  type    = "HC1",
  cluster = "group"
)

coeftest(random, vcov = vcov_cluster)

# UJI SERENTAK (WALD TEST ROBUST)
# Ambil koefisien (tanpa intercept)
b <- coef(random)[-1]

# Ambil matriks kovarians (tanpa intercept)
V <- vcov_cluster[-1, -1]

# Statistik Wald (Chi-square)
W_stat <- as.numeric(t(b) %*% solve(V) %*% b)

# df
df_wald <- length(b)
p_value <- 1 - pchisq(W_stat, df_wald)
cat("Wald Chi-square  :", W_stat, "\n")
cat("Degrees of Freedom:", df_wald, "\n")
cat("P-value          :", p_value, "\n")
```
